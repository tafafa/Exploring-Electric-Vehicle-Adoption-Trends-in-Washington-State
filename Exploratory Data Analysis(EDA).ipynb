## Univariate Analysis

```python
#setting style and color palette of the plots  for visualization
sns.set(style= "whitegrid")
sns.set_palette("crest")



#create a 3x2 subplot grid
fig, axes =plt.subplots(2, 2, figsize= (10, 6))

#plot the distribution of top 10 most common model
# Get the top 10 most common models
top_10_models = df['Model'].value_counts().head(10).index

# Filter the dataframe to include only those models
top_10_df_models = df[df['Model'].isin(top_10_models)]

sns.countplot(data=top_10_df_models, y="Model", order=top_10_models, ax=axes[0, 0])
axes[0, 0].set_title("Distrubution of Top 10 Most Common Model")
axes[0, 0].set_ylabel("Model")
axes[0, 0].set_xlabel("Count")
axes[0, 0].tick_params(axis='x', rotation=45)

#plot the distribution of top 10 most common make
# Get the top 10 most common make
top_10_makes = df['Make'].value_counts().head(10).index

# Filter the dataframe to include only those make
top_10_df_makes = df[df['Make'].isin(top_10_makes)]

sns.countplot(data=top_10_df_makes, y="Make", order=top_10_makes, ax=axes[0, 1])
axes[0, 1].set_title("Distrubution of Top 10 Most Common Make")
axes[0, 1].set_ylabel("Make")
axes[0, 1].set_xlabel("Count")
axes[0, 1].tick_params(axis='x', rotation=45)

#plot the distribution of top 10 counties
# Get the top 10 most common county
top_10_counties = df['County'].value_counts().head(10).index

# Filter the dataframe to include only those county
top_10_df_counties = df[df['County'].isin(top_10_counties)]

sns.countplot(data=top_10_df_counties, y="County", order=top_10_counties, ax=axes[1, 0])
axes[1, 0].set_title("Distrubution of Top 10 Counties")
axes[1, 0].set_ylabel("County")
axes[1, 0].set_xlabel("Count")
axes[1, 0].tick_params(axis='x', rotation=45)

#plot the distribution of top 10 cities
# Get the top 10 most common cities
top_10_cities = df['City'].value_counts().head(10).index

# Filter the dataframe to include only those cities
top_10_df_cities = df[df['City'].isin(top_10_cities)]

sns.countplot(data=top_10_df_cities, y="City", order=top_10_cities, ax=axes[1, 1])
axes[1, 1].set_title("Distrubution of Top 10 Cities")
axes[1, 1].set_ylabel("City")
axes[1, 1].set_xlabel("Count")
axes[1, 1].tick_params(axis='x', rotation=45)

#Adjust Layout
plt.tight_layout()

#show the plots
plt.show()
```

## Bivariate Analysis

```python
#EV type vs CAFV eligibility
cross_tab=pd.crosstab(df['Electric Vehicle Type'], df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'])

plt.figure(figsize=(18,6))
sns.heatmap(cross_tab, annot=True, cmap="crest", fmt='d')

plt.title('EV Type vs CAFV Eligibility', fontsize=14, pad=15)
plt.xlabel('CAFV Eligibility', fontsize=10)
plt.ylabel('EV Type', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
#Distribution of electric range by make

#Filter the data to exclude non-reported Electric Range values (0)
df_range_filtered = df[df['Electric Range'] > 0].copy()

#Filter to include ALL makes with at least 100 non-zero range registrations
# 1. Get counts of makes in the filtered data
make_counts_filtered = df_range_filtered['Make'].value_counts()
# 2. Identify makes that meet the threshold (e.g., > 100 registrations with reported range)
makes_to_plot = make_counts_filtered[make_counts_filtered >= 100].index.tolist()
# 3. Filter the range-filtered DataFrame down to these makes
df_filtered_all_makes = df_range_filtered[df_range_filtered['Make'].isin(makes_to_plot)].copy()

plt.figure(figsize=(10,10))

sns.boxplot(data=df_filtered_all_makes,y='Make',x='Electric Range',palette='crest', hue="Make", order=makes_to_plot)
plt.title('Distribution of Electric Range by Make', fontsize=14, pad=15)
plt.xlabel('Electric Range (miles)', labelpad=10)
plt.ylabel('Make', labelpad=10)

plt.tight_layout()
plt.show()
```

```python
#Electric range by model year
#Group by Model Year and get the mean Electric Range
yearly_range = df.groupby('Model Year')['Electric Range'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(data=yearly_range,x='Model Year',y='Electric Range',marker='o',color='teal',linewidth=2)
plt.title('Average Electric Range by Model Year', fontsize=14, pad=15)
plt.xlabel('Model Year', labelpad=10)
plt.ylabel('Average Electric Range (miles)', labelpad=10)


plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
